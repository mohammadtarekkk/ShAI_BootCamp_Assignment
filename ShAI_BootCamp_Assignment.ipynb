## Data Cleaning (Task 3)

```python
# make column "Id" index.

df = df.set_index("Id")
```

```python
# knowing the unique values of the column.

df["Notes"].unique()
```

```python
# knowing the unique values of the column.

df["Agency"].unique()
```

```python
# knowing the unique values of the column.

df["Status"].unique()
```

```python
# drop all columns we do not need.

df = df.drop(columns=["Notes","Status","Agency"])
```

```python
df
```

```python
# drop raws that have null values in "OvertimePay" and "BasePay" columns.

df = df.dropna(subset=["OvertimePay","BasePay"])
```

```python
df.info()
```

```python
# handle the value names in the columns.

df["EmployeeName"] = df["EmployeeName"].str.title()
df["JobTitle"] = df["JobTitle"].str.title()
```

### i will try to handle the missing values in "Benefits" column.
### there is a relation between "Benefits" and "BasePay", i will get the mean of the ratio between columns and try to fill the nan values by the same ratio.

```python
# make another dataframe without "Benefits" null values.

df2 = df.dropna(subset="Benefits")
df2
```

```python
# get the mean of the ratio between two column in variable "ratio".

ratio_array = df2["BasePay"]/df2["Benefits"]
ratio = ratio_array.mean()
print(ratio_array)
```

```python
# fillna by the same ratio.

df["Benefits"] = df.apply(lambda row : row["BasePay"] / ratio if pd.isna(row["Benefits"]) else row["Benefits"],axis=1)
df.info()
```

```python
# make "TotalPayBenefits" values correct.

df["TotalPayBenefits"] = df["Benefits"] + df["TotalPay"]
df.sample(10)
```

## Basic Data Visualization (Task 4)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPay'], bins=30, kde=True)
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
department_proportions = df['JobTitle'].value_counts().nlargest(10)
plt.pie(department_proportions, labels=department_proportions.index, autopct='%1.1f%%')
plt.title('Proportion of Employees in Different Departments')
plt.show()
```
